```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def first_degree_polynomial(x):
    matrix_A = np.array([[1, 1], [1, 15]], dtype=float)
    vector_b = np.array([func(1.), func(15.)])
    
    vector_x = linalg.solve(matrix_A, vector_b)

    return vector_x[0] + vector_x[1]*x
```

```python
x_coord = (range(1, 16))

_, ax = plt.subplots(figsize=(8,6))
ax.plot(x_coord, [func(x) for x in x_coord], label = 'Math_function')
ax.plot(x_coord, [first_degree_polynomial(x) for x in x_coord], label = 'first_degree_polynomial')
ax.set_xlabel(u'x',{'fontname':'Arial', 'size': 20})
ax.set_ylabel(u'f(x)',{'fontname':'Arial', 'size': 20})
ax.legend()

plt.show()
```

```python
def second_degree_polynomial(x):
    matrix_A = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]], dtype=float)
    vector_b = np.array([func(1.), func(8.), func(15.)])
    
    vector_x = linalg.solve(matrix_A, vector_b)

    return vector_x[0] + vector_x[1]*x + vector_x[2]*np.power(x, 2)
```

```python
x_coord = (range(1, 16))

_, ax = plt.subplots(figsize=(8,6))
ax.plot(x_coord, [func(x) for x in x_coord], label = 'Math_function')
ax.plot(x_coord, [second_degree_polynomial(x) for x in x_coord], label = 'second_degree_polynomial')
ax.set_xlabel(u'x',{'fontname':'Arial', 'size': 20})
ax.set_ylabel(u'f(x)',{'fontname':'Arial', 'size': 20})
ax.legend()

plt.show()
```

```python
def third_degree_polynomial(x=0):
    matrix_A = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]], dtype=float)
    vector_b = np.array([func(1.), func(4.), func(10.), func(15.)])
    
    vector_x = linalg.solve(matrix_A, vector_b)

    return vector_x[0] + vector_x[1]*x + vector_x[2]*np.power(x, 2) + vector_x[3]*np.power(x, 3), vector_x
```

```python
x_coord = (range(1, 16))

_, ax = plt.subplots(figsize=(8,6))
ax.plot(x_coord, [func(x) for x in x_coord], label = 'Math_function')
ax.plot(x_coord, [third_degree_polynomial(x)[0] for x in x_coord], label = 'third_degree_polynomial')
ax.set_xlabel(u'x',{'fontname':'Arial', 'size': 20})
ax.set_ylabel(u'f(x)',{'fontname':'Arial', 'size': 20})
ax.legend()

plt.show()
```
